```python
import numpy as np
import glob   

tags_path = '/Users/sjx/Desktop/CS5785/final/data/tags_train/*.txt'
files = sorted(glob.glob(tags_path))   
tags = []
for file in files:     
    f = open(file, 'r')  
    lines = f.readlines()
    for line in lines:
        tag = line.strip("\n").split(":")[1]
        if tag not in tags:
            tags.append(tag)
    f.close()
print(tags)
print(len(tags))
```

```python
from pprint import pprint
import string, re
descs_path = '/Users/sjx/Desktop/CS5785/final/data/descriptions_train/*.txt'
files = sorted(glob.glob(descs_path))
descs = []
regex = re.compile('[%s]' % re.escape(string.punctuation))
for file in files:
    f = open(file, 'r')
    lines = f.readlines()
    words = []
    for line in lines:
#         words += line.lower().translate(None, string.punctuation).split(" ")
        words += regex.sub('', line.strip("\n").lower()).split(' ')
    f.close()
    words = list(filter(None, words))
    descs.append(words)
pprint(descs[:10])
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

lmtzr = WordNetLemmatizer()
descs = [[lmtzr.lemmatize(word, "v") for word in desc] for desc in descs]
print('lemmatized:')
pprint(descs[:10])
```

```python
from nltk.corpus import stopwords
descs = [[word for word in desc if word not in stopwords.words('english')] for desc in descs]
print("filtered:")
pprint(descs[:10])
```

```python

# don't do this

from nltk import word_tokenize, pos_tag

threshold = 2

for desc in descs:
    tokens = word_tokenize(' '.join(desc))
    wordtags = pos_tag(tokens)
    nouns = [word for word, pos in wordtags if (pos == 'NN')]
    word_dict = {}
    for word in nouns:
        if word not in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] += 1
            if word_dict[word] > threshold:
                tags.append(word)
tags = list(sorted(set(tags)))
print(len(tags))
print(tags)
```

```python
kwf = np.zeros((10000, 80))
for i, desc in enumerate(descs):
    kwf[i] = np.in1d(np.asarray(tags), np.asarray(desc)).astype(int)
    print(sum(kwf[i]))
print(kwf.shape)
```
